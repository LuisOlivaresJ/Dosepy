```python
# Load dose map

## Get current directory
import os 
from pathlib import Path

import numpy as np
import plotly.express as px

from Dosepy.image import ArrayImage
from Dosepy.tools.gamma import chi

dir_path = Path(os.getcwd())

film_path = dir_path.parent / "unit" / "fixtures" / "GAMMA" / "film_dose_map_Dosepy.csv"
dcm_path = dir_path.parent / "unit" / "fixtures" / "GAMMA" / "dcm_dose_map_Dosepy.csv"

film = ArrayImage(np.genfromtxt(film_path, delimiter=","), dpi=25.4)
dcm = ArrayImage(np.genfromtxt(dcm_path, delimiter=","), dpi=25.4)
```

```python
# Calculate gamma (3%, 2mm)

g, p = film.gamma2D(
    reference=dcm,
    dose_ta=3,
    dist_ta=2,
    dose_threshold=10,
)

print(p)
px.imshow(g, zmax=3)
```

```python
# Calculate chi

chi_map, chi_rate = chi(
    reference_image=dcm,
    comparison_image=film,
    doseTA=3,
    distTA=2,
    threshold=10
)

print(chi_rate)
px.imshow(chi_map, zmax=3)
```
