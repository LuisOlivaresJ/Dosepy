```python
from Dosepy.image import load
from Dosepy.tools.equate import crop_using_ref_position
```

```python
film_path = "/media/luis/TOMO/Tiff Films/EBT4/lattice_isle/tiff/Isle_75dpi20250218_002.tif"
dicom_path = "/media/luis/TOMO/Tiff Films/EBT4/lattice_isle/RD_Axial_20x20_1mm_Acuros_4arcos.dcm"
```

```python
film = load(film_path)
```

```python
dicom = load(dicom_path)
```

```python
film.flipud()
```

```python
%matplotlib qt
film.plot()
```

```python
row_film = 512
column_film = 452
```

```python
import matplotlib.pyplot as plt

plt.imshow(dicom.array)
```

```python
row_dicom = int(dicom.shape[0]/2)
colum_dicom = int(dicom.shape[1]/2)
```

```python
film_result, dicom_result = crop_using_ref_position(
    img_film = film,
    img_dicom = dicom,
    row_film=row_film,
    column_film = column_film,
    row_dicom = row_dicom,
    column_dicom=colum_dicom,
)
```
