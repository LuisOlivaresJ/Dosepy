### Dosepy usage for an image used by FilmQA

```python
from image import load
import matplotlib.pyplot as plt

from skimage.morphology import closing, square, erosion
from skimage.measure import label, regionprops

from skimage.filters import threshold_otsu
from skimage.color import rgb2gray

import matplotlib.patches as mpatches

import numpy as np
```

```python
demo_path = Path("/home/luis/Dosepy/src/Dosepy") / "data" / "PRUEBA_LUIS.tif"
```

```python
cal_image = load(demo_path, for_calib = True)
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(8, 2.5))
ax = axes.ravel()
ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3)
```

```python
grayscale = rgb2gray(cal_image.array)
thresh = threshold_otsu(grayscale)

binary = erosion(grayscale < thresh, square(12))
#plt.imshow(binary)

fig, axes = plt.subplots(ncols=3, figsize=(8, 2.5))
ax = axes.ravel()
ax[0] = plt.subplot(1, 3, 1)
ax[1] = plt.subplot(1, 3, 2)
ax[2] = plt.subplot(1, 3, 3)

ax[0].hist(grayscale.ravel(), bins = 20)
ax[0].axvline(thresh, color='r')

ax[1].imshow(cal_image.array[:,:,0], cmap = "gray")

regions = cal_image.region_properties(crop = 5, channel = "G")

for region in regions:
    # take regions with large enough areas
    if region.area >= 10:
        # draw rectangle around segmented coins
        minr, minc, maxr, maxc = region.bbox
        rect = mpatches.Rectangle((minc, minr), maxc - minc, maxr - minr,
                                  fill=False, edgecolor='red', linewidth=1)
        ax[1].add_patch(rect)
```

```python
label_image, num = label(binary, return_num = True)
print(f"Number of images detected: {num}")
```

```python
regions = regionprops(label_image, intensity_image = cal_image.array[:,:,1])
```

```python
from skimage.filters import sobel
mean = []
i = 1
for region in regions:
    print(f"Valor medio de pixel: {region.intensity_mean}")
    region.image_intensity[region.image_intensity == 0] = np.max(region.image_intensity)
    th = threshold_otsu(region.image_intensity)
    #th = np.percentile(region.image_intensity, 40)
    #elevation_map = sobel(region.image_intensity)


    
    bin = erosion(region.image_intensity < th, square(20))
    #plt.figure()
    #plt.imshow(bin)
    fig, axes = plt.subplots(ncols=2, figsize=(8, 2.5))
    ax = axes.ravel()
    ax[0] = plt.subplot(1, 2, 1)
    ax[1] = plt.subplot(1, 2, 2)

    ax[0].hist(region.image_intensity.ravel(), bins = 20)
    ax[1].imshow(bin)
    
    ax[0].axvline(th, color='r')
    ax[0].set_title(f"Film: {i}")

    lb = label(bin)
    reg = regionprops(lb, intensity_image = region.image_intensity)
    print(reg[0].intensity_mean)
    ax[1].set_title(f"I: {reg[0].intensity_mean}")
    mean.append(reg[0].intensity_mean)

    i += 1

sorted(mean)
```
