## Primer set de películas

```python
#from Dosepy.tools.image import load
from image import load
from pathlib import Path

#Dosis usadas para irradiación
imparted_doses = [0, 0.5, 1,1.5, 2, 3, 5, 8, 10]

#Lectura de archivo
file_path_1 = Path("/home/luis/Downloads") / "PRUEBA LUIS.tif"  # Office PC
#file_path_sin = Path("/home/luis/Dosepy/src/Dosepy/data") / "PRUEBA_LUIS.tif" # Personal laptop
cal_image_1 = load(file_path_1, for_calib = True)

#Obtener curva de calibración
cal_1 = cal_image_1.get_calibration(doses = imparted_doses, channel = "G", field_in_film = True)
cal_1.plot(color = 'green')
```

## Segundo set

```python
#Lectura de archivo
file_path_2 = Path("/home/luis/Downloads") / "POS24_1.tif"  # Office PC
#file_path_sin = Path("/home/luis/Dosepy/src/Dosepy/data") / "PRUEBA_LUIS.tif" # Personal laptop
cal_image_2 = load(file_path_2, for_calib = True)

#Obtener curva de calibración
cal_2 = cal_image_2.get_calibration(doses = imparted_doses, channel = "G", field_in_film = True)
cal_2.plot(color = 'green')
```
